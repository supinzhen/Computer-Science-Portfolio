```python
# Dark - Power Law Transformation 01
 
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 讀入圖片
img_d = cv2.imread('dark.jpg', 0)
# 調整圖片大小
img_d = cv2.resize(img_d, None, fx = 0.5, fy = 0.5)
img_pow_adjust = img_d.copy()

row, col = img_d.shape

# 定義power law轉換函數
def powerLaw(p_r):
    
    # 決定c值
    p_c = 255 / (255 ** p_r)
    print(p_c)
    
    # 將圖片進行對數變換
    for i in range (0,row):
        for j in range (0,col):
            # Power Law Transformation
            img_pow_adjust[i, j] = p_c * img_d[i, j] ** p_r
    return img_pow_adjust

# 決定繪製子圖表的個數
plot_row = 3
plot_col = 3

# 定義繪製圖表函數
def plotPowerLaw(index, p_r):
    
    # 決定c值(繪製轉換曲線用)
    p_c = 255 / (255 ** p_r)
    # 進行Power Law Transformation
    result = powerLaw(p_r)
    # 設定子圖表
    plt.subplot(plot_row, plot_col, index)
    # 設定子圖表的標題
    plt.title('Image', fontsize = 10)
    # 顯示圖片
    plt.imshow(cv2.cvtColor(result, cv2.COLOR_GRAY2RGB))

    plt.subplot(plot_row, plot_col, index + 1)
    plt.title('Histogram', fontsize = 10)
    # 顯示直方圖  
    plt.hist(result.ravel(),256,[0,256])
    # 設置y軸座標
    plt.yticks([0, 5000, 10000])
    # 設置x軸座標
    plt.xticks([0,100, 200])

    # 繪製轉換曲線
    plt.subplot(plot_row, plot_col, index + 2)
    curve_name = 'Transfer Curve, r = ' + str(p_r)
    plt.title(curve_name, fontsize = 10)
    x = np.linspace(0,255)
    p_y = p_c * x ** p_r
    plt.plot(x, p_y)
    plt.yticks([0, 50, 100, 150, 200, 250])
    plt.xticks([0, 50, 100, 150, 200, 250])
    # 設定檔案名稱
    file_name = './dark/dark_pow_law_r_' + str(p_r) + '.jpg' 
    # 寫入圖片
    cv2.imwrite(file_name, cv2.cvtColor(img_pow_adjust, cv2.COLOR_GRAY2BGR))


# 繪製 r = 1 的轉換
plotPowerLaw(1, p_r = 1)
# 繪製 r = 0.67 的轉換
plotPowerLaw(4, p_r = 0.67)
# 繪製 r = 0.4 的轉換
plotPowerLaw(7, p_r = 0.4)

# 設置圖表標題
plt.suptitle('Dark - Power Law Transformation')
# 調整子圖表間距
plt.tight_layout()
# 儲存子圖表
plt.savefig('./dark/dark_power_law_01.jpg', dpi = 500)
```
